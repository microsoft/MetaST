## Notebook for pattern based analysis

The experiment is conducted on the sampled challenge set datasets shared by zhiguo. We generate the MV3 and MV4 model prediction using CSDS

```python
import pandas as pd
```

```python
MV4_model_prediction = "AllDetails.2020_02_27_14_34_48_mv4_target_set_sample_QueryPatterns_all.tsv"
MV3_model_prediction = "AllDetails.2020_02_27_14_47_04_mv3_target_set_sample_QueryPatterns_all.tsv"

df_mv4 = pd.read_csv(MV4_model_prediction, sep='\t', header=0, encoding = "utf-8")
df_mv3 = pd.read_csv(MV3_model_prediction, sep='\t', header=0, encoding = "utf-8")
```

```python
df_mv4
```

```python
df_left = pd.merge(df_mv3, df_mv4, on='Pattern', how='left',suffixes=('_mv3','_mv4'))
```

## For top patterns


```python


df_left_high = df_left[df_left['Freq_mv3']>=6]

df_left_filter = df_left_high[['Pattern','Freq_mv3','Accuracy_mv3','Accuracy_mv4']]
df_left_filter_mv3win = df_left_filter[df_left_filter['Accuracy_mv3']>df_left_filter['Accuracy_mv4']]
df_left_filter_mv4win = df_left_filter[df_left_filter['Accuracy_mv3']<df_left_filter['Accuracy_mv4']]
df_on_par = df_left_filter[df_left_filter['Accuracy_mv3']==df_left_filter['Accuracy_mv4']]
print (len(df_left_filter_mv3win),len(df_left_filter_mv4win),len(df_on_par))
df_left_filter_mv4win
```
